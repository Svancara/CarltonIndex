# C2 score based on the Calmar idea

What we want in trading? It is very simple. We want an account equity going up as much as it is possible a not going down.

This is mathematicaly: MySuccess = AccountEquity / Drawdowns

The higher the better. 

We know that formula: It is the Camar ratio: <https://www.investopedia.com/terms/c/calmarratio.asp>

However, Calmar ratio is based of 3 years timeframe. 3-years-old-systems! We, on Collective2, have a handful of such systems... 

So we need to apply the above "MySuccess" formula to much shorter systems.

The question is: Can I trust that number?
Say the 'MySuccess' number is good today. Isn't it just by accident? Was it good in the past too? If so, may be, we are lucky and we can hope that it will be good in the future too.

So we need some statistics of that number. Let's use its volatility. If the "MySuccess" number is high and it has a low volatility, we can hope we found what we are looking for.

We are looking for the highest ratio: MySuccess / Volatility 

## Quarantine

We introduce a notion of quarantine. 
Every trading strategy will be in 3 months quarantine.

Reasons: 
- Nobody should jump to too young systems.
- Statistics does not work for small number of observations. We can't say anything reliable if we do not have enough data.
- Trade leaders learn Collective2 and make mistakes at begining. (I personaly sent a short trade to the long-only system.) 
- So first months can be chaotic, but that chaos can never happen later again. (If it WILL happen later, we should find it later.)
- Systems have usually too low starting capital which does not allow to decrease risk. Strategy can easily get 50% drawdown which is shown on the system page forever. But later, when the system performs well and uses a good risk strategy, such drawdown will never happen again. In such case, it is much safer for customers to jump to such system with appropriate starting capital in their accouts. So I think it is fair not to use first chaotic numbers forever.
- Good-old-strategies doing well for a long time (maybe years), but having 50% MaxDrawdown from the first motnh are penalized in our Grid when users search systems with low drawdowns. Contrary, young systems doing well just by accident are favorized. We should do it better. Maybe we should recalculate stats every three years. I mean calculate stats just from last 3 years. (And get Calmar as it is defined. :-) ) Such 3 years stats will describe present strategies statuses, which can be very different from their youth. But their youth is gone and never returns. We should present current states verified by 3 years stats.

So: We will pretend like the systems in the quarantine do not actually exist.

We shoud place a big red text "DO NOT JUMP TO TOO YOUNG SYSTEM!" on the system page. :-)

Maybe, we should not allow to subscribe to young systems at all. Seriously. We should explain it to our customers and they should accept it.


## Statistics quarantine

We will wait another 2-3 months (40-60 trading days) for data. 
We can't do reliable statistics from 5 observations. Statistics simply does not work this way.
We will start to calculate data in this second quarantine time, but we will show no stats yet. "SORRY - WE CAN'T SAY ANYTHING RELIABLE YET! WE ARE WAITING FOR DATA!" on the system page instead of not-yet-reliable stats.

I am using 3 months for the statistics quarantine. That is 60 trading days data. 60 observations for stats.

All that means I am using only systems older than 6 months.


## Algorithm principles

Remark: I am using the word "calmar" here. But it is not calmar as it is defined. I am just using it as a description for the "profit/drawdown" principle.

- To get smooth "MySuccess" stats, we need systems which do not jump up and down. This is measured by the linear regression. Specifically, by sum of squares of the linear regression residuals adjusted by the system age. (The lower the better.)
- A linear regression line must go up.
- A strategy equity can't fall under starting capital after both quarantines.
- Calmar = CurrentProfit / ( 1 + MaxDrawDown)
- Score = Calmar / ( 1 + SD(Calmar))
- Maybe Score can be further adjusted by Calmar skew and kurtosis. This variant is using for sorting right now: 
    `ScoreAdjusted = Score * (1 + (CalmarSkew/6) - ((CalmarKurtosis-3)/24) )`

# Results

A results page contains results of the proposed score.

It shows systems selected for several dates, from today back to the history in 3 months intervals. 

Each day selections are based on combinations of the 2 score parameters:
- Calmar from 6 to 4 (descending)
- Residuals from 0.1 to 0.2 (ascending)

This setup is from more restrictive to less restrictive.

Counts of selected systems are shown too.

Detailed lists of systems selected by each combination are provided as links to details WWW pages.

The first day is for today. Its detailed page contains current C2 charts. Other pages do not contain C2 chart because it is confusing. (We would show the today's chart on the historical page which is very confusing.)
